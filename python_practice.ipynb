```python
# map() funtion apply on every iter
name=["mandeep","singh","fundds"]
age=[21,34,53,23,42]
#question 1 change name in upper case
name=list(map(lambda x:x.upper(),name))
print(name)
#question 2 calculate age after 5 year
age=list(map(lambda x:x+5,age))
print(age)
"""map() funtion return object so we need to convert it into list"""

#we use outside function also
def sub(a):
  return a-10
age=list(map(sub,age))
print(age)
```

```python
# Filter function use on every iter for filter list
val=[23,42,56,75,21,3,2,1,4]
val=list(filter(lambda x:x%2==0,val))
print(val)
"""filter() funtion return object so we need to convert it into list"""
```

```python
#Zip() funtion for combine element from multiple iterable
name=["mandeep","singh","fundds"]
age=[21,34,53,23,42]
print(*sorted(zip(name,age),key=lambda x:x[1]))
#* data unpacking
# stop at smallest list
```

```python
#Enumerate() give index and value
name=["mandeep","singh","fundds"]
for i,j in enumerate(name):
  print(i,j)
```

```python
#sort() and sorted() function for short list
#work on list
name=["mandeep","singh","fundds"]
print(sorted(name)) #sorted return new list
print(name)
name.sort() #sort change original list
print(name)

#work on dictionary
students={
    "name":["a","b","c","d"],
    "age":[21,34,53,23,42]
}
print(sorted(students['age']))
```

**Day 4 adv python practice**

```python
#Decorator example
# if function have parameter then *args and **kwargs pass
#decorator: because of if we did want change in our original fuction then we use decorator
def decorator(func):
  def wapper(*args,**kwargs):
    print("start")
    print(func(*args,**kwargs))
    print("end")
  return wapper

@decorator #by placing above a function then that function behave like a decorator
def add(a,b):
  return a+b
add(1,2)
def name(w="mandeep"):
  return w
print(name('sandeep'))
```

```python
#Expain *args and **kwargs
""" *(any name we use) because of args is only name * collect multiple positional arguments in a tuple
**(any name we use) kwargs also a name ** collect multiple keyword arguments in a dictionary
"""
def func(*val,**students):
  print(val," :",type(val))
  print(students," :",type(students))
  print(students['name'])
  print(val[1])
func(1,2,3,name="mandeep",age=23,prof="job")
# 1,2,3 is a positional argument that why they store as a tuple
#name age and prof is keyword argument
```
